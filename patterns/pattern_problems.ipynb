Pattern Problems

Every Problem -> Take as input N, a number. Print the pattern as given in the input and output section.

Problem 1 :\
Input Format : Enter value of N\
Constraints : 1<=N<10\
Output Format : Print the pattern

Sample Input :\
5

Sample Output :
``` 
1****
12***
123**
1234*
12345

````python
N = int(input())
for i in range(1,N+1):

    for j in range(1,i+1):
        print(j,end="")
    for _ in range(N-i):
        print("*",end="")
    print()
````

Problem 2 :\
Input Format : Enter value of N\
Output Format : Print the Downward Triangle Pattern

Sample Input :\
5

Sample Output :
```
* * * * * 
 * * * * 
  * * * 
   * * 
    * 


````python
n=int(input())

#Constraints aren't given in the problem so added 'if condition' -> For understanding purpose.
if(n>0):      
    for i in range(n):
        for _ in range (i):
            print(" ",end="")
    
        for _ in range(n-i):
            print("* ",end="")

        print()

else:
    print("Number isn't a positive integer!")
````

Problem 4 :\
Input Format : Enter value of N\
Constraints : N<=20\
Output Format : Print the Hollow Rhombus Pattern

Sample Input :\
7

Sample Output :
```
      *******
     *     *
    *     *
   *     *
  *     *
 *     *
*******


````python
n=int(input())

for i in range(n):

    #For first lower pyramid
    for _ in range(n-i-1):
        print(" ",end="")

    # Priting first and last row
    if i==0 or i==n-1:
        for _ in range(n):
            print("*",end="")
    
    #Printing of inner pattern
    else:
        for j in range(n):
            if j==0 or j==n-1:
                print("*",end="")  
            else: 
                print(" ",end="")

    print()
````
