Pattern 1 - Basics

Every Problem -> Take as input N, a number. Print the pattern as given in the input and output section.

Problem 1 :\
Input Format : Enter value of N\
Constraints : 1<=N<10\
Output Format : Print the pattern

Sample Input :\
5

Sample Output :\
1****\
12***\
123**\
1234*\
12345

```python
N = int(input())
for i in range(1,N+1):

    for j in range(1,i+1):
        print(j,end="")
    for _ in range(N-i):
        print("*",end="")
    print()
```
